```python
sns.barplot(x = 'City' , y = "Total" ,  data = df(), hue = 'Gender')
rcParams['figure.figsize'] = 8,5
plt.show()
```

```python
df_str_dropped = df().drop(columns=['Invoice ID','Branch', 'City','Customer type','Gender' ,'Date','Time','Payment', 'Product line' , 'gross margin percentage'])
df_str_dropped
```

```python
df_corr = df_str_dropped.corr()
df_corr
```

```python
sns.heatmap(df_corr, cmap = "gist_heat" , annot=True)
rcParams['figure.figsize'] = 10,6
plt.title('Correlation Between Numerical Data')
plt.show()
```

```python
sns.histplot(df()['gross income'] ,  bins = [0,5,10,15,20,25,30,35,40,45,50] , kde = True , color = 'y')
rcParams['figure.figsize'] = 8,5
plt.title('Plot between Gross total income and its count')
plt.show()
```

```python
sns.scatterplot(x = 'Total' , y ='gross income' , data = df() , color = 'g')
rcParams['figure.figsize'] = 8,5
plt.title('Plot Between Total Cart value vs Gross income')
plt.show()
```

```python
sns.lineplot(x = 'Total' , y ='gross income' , data = df() , color = 'g')
rcParams['figure.figsize'] = 8,5
plt.title('Plot Between Total Cart value vs Gross income')
plt.show()
```

```python
sns.histplot(df()['Rating'] ,bins = [4,5,6,7,8,9,10],kde = True , color = 'g')
rcParams['figure.figsize'] = 8,5
plt.title('Plot Between Rating and its count')
plt.show() 
```

```python
sns.histplot(df()['Total'] ,kde = True , color = 'y',bins=[0,50,100,200,300,400,500,600,700,800,900,1000])
rcParams['figure.figsize'] = 8,5
plt.title('Plot Between Total Cart values and its count')
plt.show() 
```

```python
gender_countplot = sns.countplot(x = 'Gender' , data = df())
rcParams['figure.figsize'] = 8,5
for bars in gender_countplot.containers:
    gender_countplot.bar_label(bars)
```

```python
gender_sales = df().groupby(['Gender'] , as_index = False)['Total'].sum().sort_values(by = 'Total' , ascending = False)
gender_sales.to_csv('Files\\Genderwise Total Sales')
gender_sales
```

```python
sns.barplot(x = 'Gender' , y = "Total" , data = gender_sales , color = 'g')
rcParams['figure.figsize'] = 8,5
plt.title('Plot of Gender vs Total ')
plt.show()
```

```python
city_sales = df().groupby(['City'] , as_index=False )['Total'].sum().sort_values(by = 'Total' , ascending = False)
city_sales.to_csv('Files\\Citywise Sales Total')
city_sales
```

```python
sns.barplot(x = 'City' , y = 'Total' , data = city_sales , color = 'y')
rcParams['figure.figsize'] = 10,6
plt.show()
```

```python
product_line = df().groupby(['Product line'] ,  as_index=False )['Total'].sum().sort_values(by = 'Total' , ascending=False)
product_line.to_csv('Files\\Product line Total') 
product_line
```

```python
sns.barplot(x = 'Product line' , y = 'Total' , data = product_line , color = c_cyan)
rcParams['figure.figsize'] = 12,9
plt.title('Bargraph Between Product Categories and its Total Amount')
plt.xlabel('Product - Categories' , size= 15)
plt.ylabel('Total Amount' , size= 15)
plt.show()
```

```python
product_line_gross_income = df().groupby(['Product line'] , as_index=False)['gross income'].sum().sort_values(by = 'gross income', ascending=False)
product_line_gross_income.to_csv('Files\\product line gross income')
product_line_gross_income
```

```python
sns.barplot(x = 'Product line' , y = 'gross income' , data = product_line_gross_income , color = c_blue)
rcParams['figure.figsize'] = 12,9
plt.title('Bargraph Between Product Categories and Total Gross Income')
plt.xlabel('Product - Categories' , size= 15)
plt.ylabel('Total Gross Income' , size= 15)
plt.show()
```

```python
customer_type_gross_income = df().groupby(['Customer type'] ,  as_index=False)['gross income'].sum().sort_values(by = 'gross income', ascending=False)
customer_type_gross_income.to_csv('Files\\customer type gross income')
customer_type_gross_income
```

```python
sns.barplot(x = 'Customer type' , y ='gross income' , data = customer_type_gross_income , color = c_yellow)
rcParams['figure.figsize'] = 8,6
plt.title('Bargraph Between Customer type and total gross income')
plt.show()
```

```python
gender_gross_income = df().groupby(['Gender'] , as_index=False)['gross income'].sum().sort_values(by = 'gross income' , ascending=False)
gender_gross_income
```

```python
sns.barplot(x = 'Gender' , y ='gross income' , data = gender_gross_income , color = c_green)
rcParams['figure.figsize'] = 8,6
plt.title('Bargraph Between Gender and total gross income')
plt.show()
```
